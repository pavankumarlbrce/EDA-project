```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r'D:\DSINTERNSHIP\data.xlsx.csv')
columns=data.columns
data.info()
```

```python
print(data.shape)
data['Specialization'].value_counts()
newdata=data[data['Specialization']=='computer science & engineering']
print(newdata.shape)
print(newdata['Specialization'].value_counts())
print(newdata['Designation'].value_counts())
```

```python
x=data['Specialization']
y=data['Gender']
cross_tab = pd.crosstab(x, y)
plt.bar(y,x, label='Group 1')
plt.show()
print(cross_tab)
```

```python
# print(data[['Designation','Gender']])
grouped_data = data.groupby(['Gender','Designation']).size().unstack(fill_value=0)

# Plot stacked bar chart
grouped_data.plot(kind='bar', stacked=True, figsize=(20, 20))

plt.xlabel('Gender')
plt.ylabel('Designation')
plt.title('Gender Distribution by Designation')
plt.legend(title='Gender', loc='upper right')
# plt.legend(bbox_to_anchor =(0.75, 1.15), ncol = 2)

# plt.xticks(rotation=45, ha='right')
# plt.tight_layout()

plt.show()
```
