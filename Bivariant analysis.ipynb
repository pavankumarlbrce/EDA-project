```python
import matplotlib.pyplot as plt

# Assuming 'data' and other variables are properly defined
grouped_data = data.groupby(['JobCity','Designation']).size().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(25, 20))  # Create figure and axis objects

# Plot grouped data as a stacked bar plot using the same axis object
grouped_data.plot(kind='bar', stacked=True, ax=ax)

# Rotate x-axis labels by 45 degrees
ax.tick_params(axis='x', rotation=45, labelrotation=45)

# Adjust bottom margin to prevent clipping of x-axis labels
plt.gcf().subplots_adjust(bottom=0.15)

# Set labels and title
plt.xlabel('Designation')
plt.ylabel('Gender')
plt.title('JobCity Distribution by Designation')

plt.show()
```
